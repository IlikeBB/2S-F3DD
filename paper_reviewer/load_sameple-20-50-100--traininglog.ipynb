```python
import matplotlib.pyplot as plt
import pickle
```

```python
s_count = 50
with open(f'../checkpoints/S1-segment/history_log/histroy-fold_2022.04.07-original_3T_NL-data_FTL-{s_count}--sample', 'rb') as f:
     data = pickle.load(f)
```

```python
data.keys()
```

```python
plt.plot(data['dice_coef'], label='Train Dice')
plt.plot(data['val_dice_coef'], label='Val Dice')
plt.title(f"Sample {s_count} Training + Val  - Dice Log")
plt.legend()
plt.savefig(f'../plot_results/paper_viewer/Sample {s_count} Training + Val  - Dice Log.png', dpi=600)
```

```python
plt.plot(data['IoU'], label='Train IoU')
plt.plot(data['val_IoU'], label='Val IoU')
plt.title(f"Sample {s_count} Training + Val - IoU Log")
plt.legend()
plt.savefig(f'../plot_results/paper_viewer/Sample {s_count} Training + Val  - IoU Log.png', dpi=600)
```
